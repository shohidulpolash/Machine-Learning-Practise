```python
import pandas as pd
```

```python
df1=pd.read_csv('customers.csv')
df1
```

```python
df2=pd.read_csv('orders.csv')
df2
```

# JOIN 
dataframe.merge(
    left,
    right,
    how: str = 'inner',
    on=None,
    left_on=None,
    right_on=None,
    left_index: bool = False,
    right_index: bool = False,
    sort: bool = False,
    suffixes=('_x', '_y'),
    copy: bool = True,
    indicator: bool = False,
    validate=None,
)

# Inner join 
By default dataframes join as inner join. (how='inner')

```python
df3=pd.merge(df1,df2,on='CustomerID')
```

```python
df3
```

# Outer Join

```python
df3=pd.merge(df1,df2,on='CustomerID',how='outer')
df3
```

# Left Join


```python
df3=pd.merge(df1,df2,on='CustomerID',how='left')
df3
```

# Right Join

```python
df3=pd.merge(df1,df2,on='CustomerID',how='right')
df3
```

# Indicator
For checking data available data is in both table or not

```python
df3=pd.merge(df1,df2,on='CustomerID',how='right',indicator=True)
df3
```
