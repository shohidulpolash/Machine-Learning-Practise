# Method 1

```python
x=data['text']
```

```python
y=data['label_num']
```

Applying CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
labelwords=cv.fit_transform(x)
```

```python
labelwords.toarray()
```

```python
labelwordsDF=pd.DataFrame(labelwords.toarray())
```

```python
#labelwordsDF[107].value_counts()
```

Merging CountVectorized array with main dataset

```python
data2=pd.merge(data,labelwordsDF,right_index=True,left_index=True)
```

```python
data2
```

```python
data2.drop(['text'],axis=1,inplace=True)
```

```python
data2
```

```python
x=data2.drop(['label_num'],axis=1)
```

```python
y=data2['label_num']
```

Separating testing and training data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=1)
```

Applying Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
MNB=MultinomialNB()
MNB.fit(xtrain,ytrain)
```

```python
MNB.score(xtest,ytest)
```
