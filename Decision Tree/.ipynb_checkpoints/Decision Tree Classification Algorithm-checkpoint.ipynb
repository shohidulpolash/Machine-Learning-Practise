```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
#Transform to DataFrame
df=pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])
```

```python
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x=df.drop(['target'],axis=1)
```

```python
y=df.target
```

```python
y
```

# Data Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=1)
```

# Decision Tree Classifier  (GINI IMPURITY)
DecisionTreeClassifier(
    *,
    criterion='gini',
    splitter='best',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features=None,
    random_state=None,
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    class_weight=None,
    ccp_alpha=0.0,
)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC=DecisionTreeClassifier() #by defaoult it takes GINI Impurity 
```

```python
DTC.fit(xtrain,ytrain)
```

```python
DTC.score(xtest,ytest)
```

We also can declare criterion for GINI IMPURITY.

```python
DTC1=DecisionTreeClassifier(criterion='gini')
```

```python
DTC1.fit(xtrain,ytrain)
```

```python
DTC1.score(xtest,ytest)
```

```python
predDTC1=DTC1.predict(xtest)
```

```python
predDTC1
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,predDTC1))
```

# Decision Tree Classification using  ENTROPY

```python
DTC2=DecisionTreeClassifier(criterion='entropy')
```

```python
DTC2.fit(xtrain,ytrain)
```

```python
DTC2.score(xtest,ytest)
```

```python
predDTC2=DTC2.predict(xtest)
```

```python
predDTC2
```

```python
print(classification_report(ytest,predDTC2))
```

# How to view the tree!

```python
from sklearn import tree
tree.plot_tree(DTC2) 
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(DTC2, out_file=None, 
                      feature_names=data.feature_names,  
                      class_names=data.target_names,  
                    filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
```
